```
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage import io, exposure
```

```
image1 = io.imread("parrot.jpg")
```

##Escala de grises

```
rgb_weights = [0.2989, 0.5870, 0.1140]
gray_image = np.dot(image1, rgb_weights)
plt.imshow(gray_image, 'gray')
```

##Histograma

```
plt.hist(gray_image.ravel(), bins = 256);
```

```
ax = plt.hist(gray_image.ravel(), bins = 256)
plt.show()
```

#Parte 2 (b)

```
from skimage import exposure
from skimage.color import rgb2gray

img2 = io.imread("parrot.jpg")
imagen_name = skimage.color.rgb2gray(img2)

```

##Transformacion logaritmica

```
img_log = exposure.adjust_log(imagen_name,1)
```

```
plt.imshow(img_log, 'gray')
plt.title("Transformacion log")
plt.show()
```

```
plt.hist(img_log.ravel(), bins = 256);
plt.title("Histograma transformacion log")
plt.show()
```

##Ecualizacion de histograma


```
p2, p98 = np.percentile(imagen_name, (2,98))

# estiramiento del contraste
img_rescale = exposure.rescale_intensity(imagen_name, in_range=(p2,p98))

# ecualizacion
image_eq = exposure.equalize_hist(imagen_name)

# ecualizacion adaptativa
image_adap = exposure.equalize_adapthist(imagen_name, clip_limit=0.03)

```

```
plt.imshow(imagen_name, 'gray')
plt.title("Normal")
plt.show()
```

```
plt.hist(imagen_name.ravel(), bins = 256);
plt.title("Normal")
plt.show()
```

```
plt.imshow(image_eq, 'gray')
plt.title("Ecualizacion")
plt.show()
```

```
plt.hist(image_eq.ravel(), bins = 256);
plt.title("Ecualizacion")
plt.show()
```

```
plt.imshow(image_adap, 'gray')
plt.title("Ecualizacion adaptativa")
plt.show()
```

```
plt.hist(image_adap.ravel(), bins = 256);
plt.title("Ecualizacion adaptativa")
plt.show()
```

##Transformada inversa

```
imagen_inversa = skimage.util.invert(img_eq)
```

```
plt.imshow(imagen_inversa, 'gray')
plt.title("Imagen inversa de ecualizacion")
plt.show()
```

```
plt.hist(imagen_inversa.ravel(), bins = 256);
plt.title("Histograma inversa ecualizacion")
plt.show()
```
